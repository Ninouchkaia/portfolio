# Amino Acid Usage Visualization
Visualization for PaxDB DS2.

```
import pandas as pd
import matplotlib.pyplot as plt

usage = pd.read_csv('../results/amino_acid_usage_weighted.tsv', sep='\t')
usage.head()
```

```
usage.set_index('species_id').T.plot(figsize=(10,6))
plt.ylabel('Relative Frequency')
plt.show()
```
